```python
import os 
import numpy as np
```

```python
from natsort import natsorted
import pandas as pd 
from tqdm import tqdm
folders = natsorted(os.listdir("../data/generated/"))[-64:]
dfs = []
pbar = tqdm(folders)
for folder in pbar:
  try:
    a = pd.read_csv("../data/generated/%s/data.csv"%folder, skiprows=2) 
  except:
    print("ERROR: ", folder)
    continue
  pbar.set_description(folder + str(a.shape))
  if np.prod(a.shape)>0:
    dfs.append(a.values)

df = np.concatenate(dfs, axis=0) 
print(df.shape)
```
